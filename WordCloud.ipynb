```r
install.packages("wordcloud",repos="http://cran.r-project.org")
```

```r
library(wordcloud)
```

```r
output <- read.csv("MR_Output.csv")
```

```r
colnames(output) <- c("Word","Count")
```

```r
head(output)
```

```r
output$Count <- as.numeric(output$Count)
```

```r
set.seed(1234)
wordcloud(words = output$Word, freq = output$Count,scale=c(4,.6), min.freq = 2,
          max.words=500, random.order=FALSE, rot.per=0.35, 
          colors=brewer.pal(8, "Dark2"))
```
